#Creating Synthetic Data


```python
import numpy as np
import pandas as pd

np.random.seed(42)

dates = pd.date_range(start="2025-01-01", periods=180, freq="D")

data = []

for level, mean, sd in [
    ("Level 1", 100, 2),
    ("Level 2", 250, 4)
]:
    values = np.random.normal(mean, sd, size=len(dates))

    # Add drift after day 120
    values[120:] += np.linspace(0, 6, len(values[120:]))

    for d, v in zip(dates, values):
        data.append([d, level, v])

df = pd.DataFrame(data, columns=["date", "qc_level", "value"])
df.head()
```

```python
#Save the dataset

df.to_csv("data/raw/qc_values.csv", index=False)
```
